```python
import pandas as pd
df = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')
print("Dataset loaded successfully.")
```

```python
print("DataFrame Info:")
df.info()
print("\nFirst 5 rows of the DataFrame:")
display(df.head())
```

```python
print("Missing values before handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])

initial_columns = df.shape[1]
missing_percent = df.isnull().sum() / len(df) * 100
columns_to_drop = missing_percent[missing_percent > 50].index.tolist()
if columns_to_drop:
    df = df.drop(columns=columns_to_drop)
    print(f"Dropped columns due to high missing values (>50%): {columns_to_drop}")

print("\nMissing values after dropping high-missing columns:")
display(df.isnull().sum()[df.isnull().sum() > 0])
if df.isnull().sum().any():
    df.dropna(inplace=True)
    print("Dropped rows with any remaining missing values.")

final_columns = df.shape[1]
print(f"\nColumns before handling missing values: {initial_columns}, after: {final_columns}")
print("Final missing values check:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
final_rows = df.shape[0]
print(f"Number of rows before removing duplicates: {initial_rows}")
print(f"Number of rows after removing duplicates: {final_rows}")
print(f"Number of duplicate rows removed: {initial_rows - final_rows}")
```

```python
print("Unique values for 'PRODUCTLINE':")
print(df['PRODUCTLINE'].unique())

print("\nUnique values for 'DEALSIZE':")
print(df['DEALSIZE'].unique())

print("\nUnique values for 'CUSTOMERNAME' (first 10 unique):")
print(df['CUSTOMERNAME'].unique()[:10])
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.title()
print("\n'PRODUCTLINE' after standardization (title case):")
print(df['PRODUCTLINE'].unique())
df['DEALSIZE'] = df['DEALSIZE'].str.title()
print("\n'DEALSIZE' after standardization (title case):")
print(df['DEALSIZE'].unique())
```

```python
print("Data type of 'ORDERDATE' before conversion:", df['ORDERDATE'].dtype)
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

print("Data type of 'ORDERDATE' after conversion:", df['ORDERDATE'].dtype)
print("First 5 'ORDERDATE' values after conversion:")
display(df['ORDERDATE'].head())
```

```python
print("Original column names:")
print(df.columns.tolist())
df.columns = df.columns.str.lower().str.replace(' ', '_')

print("\nNew column names:")
print(df.columns.tolist())
```

```python
print("Data types before final check:")
df.info()
df['ordernumber'] = df['ordernumber'].astype(str)
for col in ['productline', 'productcode', 'customername', 'city', 'country', 'dealsize']:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].astype('category')
for col in ['quantityordered', 'priceeach', 'sales', 'orderlinenumber']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print("\nData types after final check and conversion:")
df.info()
```
